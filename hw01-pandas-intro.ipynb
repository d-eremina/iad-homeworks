**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
import time

start = time.time()
dict_sum = {}
dict_count = {}
for i in df['race/ethnicity'].unique():
    dict_sum[i] = 0
    dict_count[i] = 0
for index, row in df.iterrows():
    dict_sum[row['race/ethnicity']] += row['reading score']
    dict_count[row['race/ethnicity']] += 1
end = time.time()

for i in sorted(dict_sum.keys()):
    print(i,dict_sum[i]/dict_count[i])
print((end - start)*1000, 'ms')
```

```python
import time

start = time.time()
dict_ = {}
for i in df['education'].unique():
    dict_[i] = 1e10
for index, row in df.iterrows():
    dict_[row['education']] = min(dict_[row['education']], row['writing score'])
end = time.time()

for i in sorted(dict_.keys()):
    print(i, dict_[i])
print((end - start)*1000, 'ms')
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
    if average_mark >= 90:
        return 'A'
    if average_mark >= 80:
        return 'B'
    if average_mark >= 70:
        return 'C'
    if average_mark >= 60:
        return 'D'
    if average_mark >= 50:
        return 'E'
    return 'F'

df['Grade'] = df.apply(lambda x: GetGrade((x['math score'] + x['reading score'] + x['writing score']) / 3), axis=1)
df.head(10)
```

```python
for i in sorted(df["Grade"].unique()):
    print(i + ":", len(df[(df["Grade"] == i)]))
```
